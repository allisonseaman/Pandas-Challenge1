```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
total = purchase_data["SN"].count()
total_df=pd.DataFrame({"Total Player":[total]})
total_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique = len(purchase_data["Item ID"].unique())
average = purchase_data["Price"].mean()
totalrevenue=purchase_data["Price"].sum()
summary_df = pd.DataFrame({"Number of Unique Items":[unique],
                           "Average Price": [average], 
                           "Total Revenue": [totalrevenue]})
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender=purchase_data.groupby("Gender")
gender_count=gender.nunique()["SN"]
gender_percent=(gender_count/total)*100
summary_df = pd.DataFrame({"Total Count":[gender_count],
                           "Percentage of Players":
                            [gender_percent]})
summary_df.index.name = None
summary_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchasecount=gender["Purchase ID"].count()
avg_price=gender["Price"].mean()
avg_total=gender["Price"].sum()
purchase_total=avg_total/gender_count
summary2=pd.DataFrame({"Purchase Count":purchasecount, "Average Purchase Price":avg_price,"Total Purchase Price":
                      avg_total, "Average Purchase Total per Person by Gender":purchase_total})
summary2
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins=[0,9.99,14.99,19.99,24.99,29.99,34.99,39.99,99999]
label=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
purchase_data["Age Group"]=pd.cut(purchase_data["Age"],bins,labels=label)
new=purchase_data.groupby("Age Group")
count=new["SN"].nunique()
percentage=(count/total)*100
summary3=pd.DataFrame({"Percentage of Players":percentage,
                        "Total Count":count})
summary3                        
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_count=new["Purchase ID"].count()
avgprice=new["Price"].mean()
totalpurchasevalue=new["Price"].sum()
avg_per_person=(totalpurchasevalue/count)*100
summary3=pd.DataFrame({"Purchase Count":purchase_count,
                      "Average Purchase Price": avgprice,
                      "Total Purchase Value": totalpurchasevalue,
                      "Average Purchase Total per Person":avg_per_person})
summary3
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spenders=purchase_data.groupby("SN")
purchase_count=top_spenders["Purchase ID"].count()
avg_purchase_price=top_spenders["Price"].mean()
total_purchase_value=top_spenders["Price"].sum()
summary4=pd.DataFrame({"Purchase Count":purchase_count,
                      "Average Purchase Price":avg_purchase_price,
                      "Total Purchase Value": total_purchase_value})
summary4.sort_values(["Purchase Count"],ascending=0)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_items=purchase_data[["Item ID", "Item Name", "Price"]]
items=popular_items.groupby(["Item ID", "Item Name"])
purchase_count=items["Price"].count()
value=(items["Price"].sum())
item_price=value/purchase_count
summary5=pd.DataFrame({"Purchase Count":purchase_count,
                      "Item Price":value,
                      "Total Purchase Value": value})
summary5.sort_values(["Purchase Count"],ascending=0)
```
